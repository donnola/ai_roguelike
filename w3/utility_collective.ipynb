```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def attack_enemy(base_dist, enemy_dist):
  return (30 - base_dist * 11) * 15 - 100 * enemy_dist

img = np.zeros((8, 8))

for x in np.arange(0, 8, 1):
  for y in np.arange(0, 8, 1):
    img[x][y] = attack_enemy(x, y)

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(13, 5)
im1 = ax[0].imshow(img, extent=[0, 8, 8, 0], cmap='Greys')
ax[0].set_xlabel("enemy_dist")
ax[0].set_ylabel("base_dist")
plt.colorbar(im1)

im2 = ax[1].imshow(img > 100, extent=[0, 8, 8, 0], cmap='Greys')
ax[1].set_xlabel("enemy_dist")
ax[1].set_ylabel("base_dist")
plt.colorbar(im2)
```

```python
def move_to_base(base_dist, teammate_dist):
    return base_dist * 20 + 20 * teammate_dist

img = np.zeros((8, 8))

for x in np.arange(0, 8, 1):
  for y in np.arange(0, 8, 1):
    img[x][y] = move_to_base(x, y)

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(13, 5)
im1 = ax[0].imshow(img, extent=[0, 8, 8, 0], cmap='Greys')
ax[0].set_xlabel("teammate_dist")
ax[0].set_ylabel("base_dist")
plt.colorbar(im1)

im2 = ax[1].imshow(img > 100, extent=[0, 8, 8, 0], cmap='Greys') ## random_move utility = 100
ax[1].set_ylabel("base_dist")
plt.colorbar(im2)
```
